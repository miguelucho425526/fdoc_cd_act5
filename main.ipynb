```
import pandas as pd
from google.colab import files

# Recreate df and actividad_semana5.csv as the kernel state may have reset
datos = {
    'Nombre':  ['Ana', 'Bob', 'Clara', 'Diego', 'Eva'],
    'Edad':    [25, 30, 22, None, 28],
    'Ciudad':  ['Madrid', 'Lima', 'Bogotá', 'Medellín', None],
    'Ingreso': [3000, 4500, 2800, 5000, None]
}
df = pd.DataFrame(datos)
df.to_csv('actividad_semana5.csv', index=False)

# Recreate seleccion_columnas.csv (from cell JoyAMKGbq5Sw's logic, using the freshly created df)
df[['Nombre', 'Edad', 'Ciudad']].to_csv('seleccion_columnas.csv', index=False)

# Recreate personas_filtradas.csv (from cell Rlep-SnlrBko's logic, using the freshly created df)
df_integrado = df.copy()
df_integrado['Edad'] = df_integrado['Edad'] + 5 # Apply the age increment
df_integrado['Ingreso_anual'] = df_integrado['Ingreso'].fillna(0) * 12 # Calculate annual income, handling nulls

filtro = df_integrado['Ingreso_anual'] > 36000
df_filtrado = df_integrado[filtro].sort_values('Ingreso_anual', ascending=False)
df_filtrado.to_csv('personas_filtradas.csv', index=False)

# 2. CSV generados
files.download('actividad_semana5.csv')
files.download('seleccion_columnas.csv')
files.download('personas_filtradas.csv')

print("¡Todos los archivos descargados!")
```
