```python

import pandas as pd
from google.colab import files
import io

df = pd.read_csv('actividad_semana5.csv')

print("Dataset original cargado correctamente:")
display(df)
```

```python

df_seleccion = df[['Nombre', 'Edad', 'Ciudad']]

df_seleccion.to_csv('seleccion_columnas.csv', index=False)

print("Ejercicio 1 completado → seleccion_columnas.csv generado")
display(df_seleccion)
```

```python

df_transformado = df.copy()


df_transformado['Edad'] = df_transformado['Edad'] + 5

df_transformado['Ingreso_anual'] = df_transformado['Ingreso'].fillna(0) * 12

filtro = df_transformado['Ingreso_anual'] > 36000
df_filtrado = df_transformado[filtro]

df_filtrado = df_filtrado.sort_values('Ingreso_anual', ascending=False)

    
df_filtrado.to_csv('personas_filtradas.csv', index=False)

print("Ejercicio 2 completado → personas_filtradas.csv generado")
display(df_filtrado)
```
